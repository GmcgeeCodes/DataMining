```python
def true_positives(colnum):
    tp = 0
    if(colnum > 0 and colnum < 1411):
       for i in range (162): 
         if((df.iloc[i,0][0] == 'C') and (df.iloc[i, colnum] == 1.0)):
               tp+=1
    return tp
```

```python
def false_positives(colnum):
    fp = 0
    if(colnum > 0 and colnum < 1411):
       for i in range (162): 
         if((df.iloc[i,0][0] == 'N') and (df.iloc[i, colnum] == 1.0)):
               fp+=1
    return fp
```

```python
def true_negatives(colnum):
    tn = 0
    if(colnum > 0 and colnum < 1411):
       for i in range (162): 
         if((df.iloc[i,0][0] == 'N') and (df.iloc[i, colnum] == 0.0)):
               tn+=1
    return tn
```

```python
def false_negatives(colnum):
    fn = 0
    if(colnum > 0 and colnum < 1411):
       for i in range (162): 
         if((df.iloc[i,0][0] == 'C') and (df.iloc[i, colnum] == 0.0)):
               fn+=1
    return fn
```

```python
def is_truePos(rowNum, colNumber):
    if((df.iloc[rowNum,0][0] == 'C') and (df.iloc[rowNum, colNumber] == 1.0)):
        return True 
    else:
        return False
    
def is_trueNeg(rowNum, colNumber):
    if((df.iloc[rowNum,0][0] == 'N') and (df.iloc[rowNum, colNumber] == 0.0)):
        return True 
    else:
        return False

def is_falseNeg(rowNum, colNumber):
    if((df.iloc[rowNum,0][0] == 'C') and (df.iloc[rowNum, colNumber] == 0.0)):
        return True 
    else:
        return False
    
def is_falsePos(rowNum, colNumber):
    if((df.iloc[rowNum,0][0] == 'N') and (df.iloc[rowNum, colNumber] == 1.0)):
        return True 
    else:
        return False
```

```python
ABdf = pd.DataFrame(columns = ['Group A', 'Group B'])
```

```python
for i in range(162):
    if is_truePos(i ,107):
       ABdf = ABdf.append({'Group A':df.iloc[i, 0]}, ignore_index = True)
        
    elif is_trueNeg(i, 107):
       ABdf = ABdf.append({'Group B':df.iloc[i, 0]}, ignore_index = True)
    elif is_falseNeg(i, 107):
         ABdf = ABdf.append({'Group B':df.iloc[i, 0]}, ignore_index = True)
    else:
         ABdf = ABdf.append({'Group A':df.iloc[i, 0]}, ignore_index = True)
        
        
```
